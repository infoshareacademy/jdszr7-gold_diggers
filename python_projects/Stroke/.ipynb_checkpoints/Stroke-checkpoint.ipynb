```python
# GridSearchCV?
```

```python
drzewo = DecisionTreeClassifier(random_state = 42)

param_drzewo = {"max_depth": [3, 4, 5, 6, 7],
                "criterion": ['entropy', 'gini'],
              "max_features": list(range(3,10)),
                "class_weight": [{0: 1, 1: 1}, {0: 1, 1: 5}, {0: 2, 1: 5}, {0: 1, 1: 10},{0: 1, 1: 15}, {0: 2, 1: 15}]
             }
```

```python
#grid search definition
def grid_search(classifier, param_grid, scoring, cross_val = 0):
    if cross_val == 0:
        return GridSearchCV(classifier, param_grid=param_grid, scoring = scoring, verbose=0)
    if cross_val == 1:
        return GridSearchCV(classifier, param_grid=param_grid, scoring = scoring, verbose=0)
    if cross_val >= 2:
        return GridSearchCV(classifier, param_grid=param_grid, scoring = scoring, verbose=0, cv = cross_val)

# grid_search.fit(X, y)
```

```python
cross_val = [0]

for cv_value in cross_val:
    for scorer in ['f1','recall']:
        # grid search
        grid_search_clf = grid_search(drzewo, param_drzewo, scorer, cv_value)
        grid_search_clf.fit(X_train_scaled, y_train) # w drzewie nie muszą być dane zestandaryzowanne
        best_est = grid_search_clf.best_estimator_

        #predykcja dla najlepszego estymatora wg recall
        y_pred = best_est.predict(X_test)
        y_pred_proba = best_est.predict_proba(X_test)[:,1]
        #wartość precyzji dla najlepszego estymatora
        precision = precision_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)

        precision_c, recall_c, thresholds = precision_recall_curve(y_test, y_pred_proba)


        print(f"Liczba kroswalidacji: {cv_value} \n \
Sprawdzana metryka: {scorer} \n \
Najlepsza wartość metryki: {grid_search_clf.best_score_} \n \
Precision dla najlepszego estymatora: {precision} \n \
F1_score dla najlepszego estymatora: {f1} \n \
Najlepsze parametry: {grid_search_clf.best_params_} \n \
Macierz konryngencji:")
        print(confusion_matrix(y_test, y_pred))
        plt.figure(figsize=(15, 10))
        plt.plot(recall_c, precision_c)
        plt.xlabel("Recall")
        plt.ylabel("Precision")
        print(classification_report(y_test, y_pred))
        
        
# UWAGA! Nie wiem, czemu scorer = 'f1' i f1_score(y_test, y_pred) dają inne wartości nawet bez walidacji krzyżowej
# (czyli nie liczymy średniej z n wyników). Może to jest uśredniona wartość z klas 0 i 1,
# a f1_score(y_test, y_pred) podaje tylko f1 dla jedynek (pokrywa się to z classification_report)
```

### Oversampling

```python
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
```

```python
drzewo = DecisionTreeClassifier(criterion = 'entropy', max_depth = 7)
```

```python
ros = RandomOverSampler(random_state=42)
X_train_ros, y_train_ros= ros.fit_resample(X_train, y_train)
```

```python
# liczebnośc klas
print(sorted(Counter(y_train_ros).items()))
```

```python
drzewo_ros = drzewo.fit(X_train_ros, y_train_ros)
drzewo_ros_pred = drzewo_ros.predict(X_test)
```
