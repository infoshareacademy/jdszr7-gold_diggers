```python
import pandas as pd
```

```python
df = pd.read_csv('ks-projects-201801.csv')
```

```python
#czyszczenie danych
df_clean = df.copy()
```

```python
df_clean[df_clean.country == 'N,0"']
```

```python
#usuniecie kolumn z country N,0"
df_clean = df_clean.drop(df_clean[df_clean.country == 'N,0"'].index)
```

```python
df_clean[df_clean.country == 'N,0"']
```

```python
df_clean.state.unique()
```

```python
df_clean = df_clean.drop(df_clean.query('state == "canceled" | state == "live" | state == "suspended" ').index)
```

```python
#zmiana typow danych
df_clean.main_category = df_clean.main_category.astype('category')
df_clean.currency = df_clean.currency.astype('category')
df_clean.state = df_clean.state.astype('category')
df_clean.country = df_clean.country.astype('category')
df_clean.launched = pd.to_datetime(df_clean.launched)
df_clean.deadline = pd.to_datetime(df_clean.deadline)
```

```python
df_clean.info()
```

```python
#usuniecie rekordow z data 1970
df_clean = df_clean.drop(df_clean.query('launched.dt.year < 1999').index)
```

```python
df_clean.query('launched.dt.year < 1999')
```

```python
# tworze kolumny z rokiem i miesiacem rozpoczynania zbiorek
df_clean['launched_month'] = df_clean.launched.dt.month
df_clean['launched_year'] = df_clean.launched.dt.year
```

```python
df_clean.info()
```

```python
df_clean.groupby(['launched_year', 'launched_month'])['ID'].count().reset_index()
```

```python
#wyrzucenie danych z 2018 roku
df_clean = df_clean.drop(df_clean.query('launched_year == 2018').index)
```

```python
df_clean.country.value_counts()
```

```python
df_clean.drop(['goal','pledged', 'usd pledged'], axis=1, inplace=True)
```

```python
df_clean.info()
```

```python
df_clean.state.unique()
```

```python
df_clean['launched_date'] = pd.to_datetime(df_clean['launched']).dt.date
```

```python
df_clean['launched_date'] = pd.to_datetime(df_clean['launched_date'])
```

```python
df_clean['duration'] = df_clean['deadline'] - df_clean['launched_date']
```
